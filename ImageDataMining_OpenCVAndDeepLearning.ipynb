# Image Data Mining:Open CV And Deep Learning

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate color histogram for images
def calculate_color_histogram(image_path):
    # Read the image
    image = cv2.imread(image_path)
    # Convert image to RGB (OpenCV uses BGR by default)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Calculate the histogram
    histogram = [cv2.calcHist([image], [i], None, [256], [0, 256]) for i in range(3)]
    return histogram
```

```python
# Paths to the images
image_paths = ['/content/SJSUOfficers2019.jpeg', '/content/SJSU_Project_Expo.jpeg']
```

```python
# Load images
image1 = cv2.imread('/content/SJSUOfficers2019.jpeg')
image2 = cv2.imread('/content/SJSU_Project_Expo.jpeg')

# Convert the images from BGR to RGB because OpenCV loads images in BGR
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# Display the images
plt.figure(figsize=(14, 7))

# Display Image 1
plt.subplot(1, 2, 1)
plt.imshow(image1_rgb)
plt.title('Image 1')
plt.axis('off')

# Display Image 2
plt.subplot(1, 2, 2)
plt.imshow(image2_rgb)
plt.title('Image 2')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Calculate histograms for each image
histograms = [calculate_color_histogram(path) for path in image_paths]
```

```python
# Plot the histograms
plt.figure(figsize=(14, 5))

for i, hist in enumerate(histograms):
    plt.subplot(1, len(histograms), i + 1)
    for c, color in enumerate(['r', 'g', 'b']):
        plt.plot(hist[c], color=color)
        plt.xlim([0, 256])
    plt.title(f'Color Histogram for Image {i+1}')

plt.tight_layout()
plt.show()
```
